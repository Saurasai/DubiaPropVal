```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from CSV file
df = pd.read_csv('../data/data_science_challenge_data.csv')

# Display first few rows to understand the structure and get a glimpse of the data
print("----- Head -----")
print(df.head())

# Show info about data types, non-null counts, and memory usage
print("\n----- Info -----")
print(df.info())

# Get statistical summary of numerical columns like count, mean, std, min, max, quartiles
print("\n----- Describe -----")
print(df.describe())

# Check for missing values in each column
print("\n----- Missing Values -----")
print(df.isnull().sum())

# Check if there are any duplicate rows
print(f"\n----- Duplicates: {df.duplicated().sum()}")

# Display distribution of categorical variables to understand frequency of categories
categorical_cols = ['bedrooms', 'bathrooms', 'neighbourhood', 'building']
for col in categorical_cols:
    print(f"\n----- Value counts for {col} -----")
    print(df[col].value_counts())

# Plot correlation matrix heatmap for numerical features to check feature relationships
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()

# Plot distribution of the target variable 'price' to see skewness or outliers
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.show()

# Scatter plot to visualize relationship between size and price
plt.figure(figsize=(8, 6))
sns.scatterplot(x='size', y='price', data=df)
plt.title('Price vs Size')
plt.show()

# Boxplot to compare price distribution across different neighbourhoods
plt.figure(figsize=(12, 6))
sns.boxplot(x='neighbourhood', y='price', data=df)
plt.xticks(rotation=90)  # Rotate labels for readability
plt.title('Price by Neighbourhood')
plt.show()

# Boxplot to analyze price variation by number of bedrooms
plt.figure(figsize=(8, 6))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Price by Number of Bedrooms')
plt.show()

# Boxplot to analyze price variation by number of bathrooms
plt.figure(figsize=(8, 6))
sns.boxplot(x='bathrooms', y='price', data=df)
plt.title('Price by Number of Bathrooms')
plt.show()

# Boxplot to detect outliers in the size feature
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['size'])
plt.title('Boxplot of Property Size')
plt.show()

# Boxplot to detect outliers in the price feature
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['price'])
plt.title('Boxplot of Property Price')
plt.show()

# Create size bins (groups) to analyze average price trends by size categories and neighbourhood
df['size_bin'] = pd.cut(df['size'], bins=10)
plt.figure(figsize=(14, 7))
sns.barplot(x='size_bin', y='price', hue='neighbourhood', data=df)
plt.xticks(rotation=45)  # Rotate labels for readability
plt.title('Average Price by Size Bins and Neighbourhood')
plt.legend(loc='upper right')
plt.show()
```
